* Use the Wallis Product to determine the value of $\pi$
$$ \pi = 2\times \frac{2\cdot2\cdot4\cdot4\cdot6\cdot6\cdots}{1\cdot3\cdot3\cdot5\cdot5\cdot7\cdots} $$

```python
n=5000
iteration=np.arange(1, n+1)
pi_array=np.zeros(n)
error=np.zeros(n)
multiply=1
for i in iteration:
    series=4*i**2/(4*i**2-1)
    multiply=multiply*series
    appro_pi=2*multiply
    er=np.abs(np.pi-appro_pi)
    pi_array[i-1]=appro_pi
    error[i-1]=er

length=np.arange(1, len(pi_array)+1)

fig, ax=plt.subplots(nrows=2, ncols=1, figsize=(15,10))

ax[0].semilogx(length, pi_array, 
               color="r", 
               linewidth=2, 
               label=r"Approximated $\pi$")
ax[0].plot([1, length[-1]], [np.pi, np.pi], 
           linestyle="--", 
           linewidth=1.5, 
           label=r"$\pi$")
ax[0].set_xlabel("Number of iteration in log scale")
ax[0].set_ylabel(r"Approximated value of $\pi$")
ax[0].set_xlim(left=10**(0))
ax[0].grid(True, which="both", ls="-", color='0.65')
ax[0].legend()
ax[1].semilogx(length, error, 
               color="DarkBlue", 
               linewidth=2)
ax[1].set_xlabel("Number of iteration in log scale")
ax[1].set_ylabel('Absolute value of error')
ax[1].set_xlim(left=10**(0))
ax[1].grid(True, which="both", ls="-", color='0.65')
```

* Use the continued fraction to determine $\pi$
$$\frac{\pi}{4} =1+\cfrac{1^2}{2+\cfrac{3^2}{2+\cfrac{5^2}{2+\cdots}}}$$

```python
 def FourByPi(its, ctr = 1):
    if its == 0:
        return 0
    else:
        x = ((2*ctr)-1)**2 / (2 + FourByPi(its - 1, ctr + 1))
        return  x + (1 if ctr == 1 else 0)
n=2000
iteration=np.arange(1, n+1)
pi_array=np.zeros(n)
error=np.zeros(n)
for i in iteration:
    appro_pi=4/FourByPi(i,1)
    er=np.abs(np.pi-appro_pi)
    pi_array[i-1]=appro_pi
    error[i-1]=er
length=np.arange(1, len(pi_array)+1)

fig, ax=plt.subplots(nrows=2, ncols=1, figsize=(15,10))

ax[0].semilogx(length, pi_array, 
               color="r", 
               linewidth=2, 
               label=r"Approximated $\pi$")
ax[0].plot([1, length[-1]], [np.pi, np.pi], 
           linestyle="--", 
           linewidth=1.5, 
           label=r"$\pi$")
ax[0].set_xlabel("Number of iteration in log scale")
ax[0].set_ylabel(r"Approximated value of $\pi$")
ax[0].set_xlim(left=10**(0))
ax[0].grid(True, which="both", ls="-", color='0.65')
ax[0].legend()
ax[1].semilogx(length, error, 
               color="DarkBlue", 
               linewidth=2)
ax[1].set_xlabel("Number of iteration in log scale")
ax[1].set_ylabel('Absolute value of error')
ax[1].set_xlim(left=10**(0))
ax[1].grid(True, which="both", ls="-", color='0.65')
```

* Use the Gregory–Leibniz series to determine the value of $\pi$
$$ \pi = 4(1-\cfrac{1}{3}+\cfrac{1}{5}-\cfrac{1}{7}+\cfrac{1}{9}-\cdots)$$

```python
n=5000
iteration=np.arange(1, n+1)
pi_array=np.zeros(n)
error=np.zeros(n)
s=0
for i in iteration:
    series=(-1)**(i+1)/(2*i-1)
    s=s+4*series
    er=np.abs(np.pi-s)
    pi_array[i-1]=s
    error[i-1]=er
length=np.arange(1, len(pi_array)+1)

fig, ax=plt.subplots(nrows=2, ncols=1, figsize=(15,10))

ax[0].semilogx(length, pi_array, 
               color="r", 
               linewidth=2, 
               label=r"Approximated $\pi$")
ax[0].plot([1, length[-1]], [np.pi, np.pi], 
           linestyle="--", 
           linewidth=2, 
           label=r"$\pi$")
ax[0].set_xlabel("Number of iteration in log scale")
ax[0].set_ylabel(r"Approximated value of $\pi$")
ax[0].set_xlim(left=10**(0))
ax[0].grid(True, which="both", ls="-", color='0.65')
ax[0].legend()

ax[1].semilogx(length, error, 
               color="DarkBlue", 
               linewidth=2)
ax[1].set_xlabel("Number of iteration in log scale")
ax[1].set_ylabel('Absolute value of error')
ax[1].set_xlim(left=10**(0))
ax[1].grid(True, which="both", ls="-", color='0.65')
```

* Use the ArcSin series to determine the value of $\pi$
$$ \pi = 6(\frac{1}{2}+\cfrac{1}{2\cdot3\cdot2^3}+\cfrac{1\cdot3}{2\cdot4\cdot5\cdot2^5}+\cfrac{1\cdot3\cdot5}{2\cdot4\cdot6\cdot7\cdot2^7}+\cdots)$$

```python

n=10
pi_array=np.zeros(n)
error=np.zeros(n)
s=0
for k in np.arange(n, dtype="int64"):
    a=factorial(2*k)/(factorial(k)*factorial(k))
    b=(16**k)*(2*k+1)
    series=3*a/b
    s=s+series
    er=np.abs(np.pi-s)
    pi_array[k]=s
    error[k]=er
length=np.arange(1,n+1)
fig, ax=plt.subplots(nrows=2, ncols=1, figsize=(15,10))

ax[0].plot(length, pi_array, 
               color="r", 
               linewidth=2, 
               label=r"Approximated $\pi$")
ax[0].plot([1, length[-1]], [np.pi, np.pi], 
           linestyle="--", 
           linewidth=2, 
           label=r"$\pi$")
ax[0].set_xlabel("Number of iteration")
ax[0].set_ylabel(r"Approximated value of $\pi$")
ax[0].set_xlim(left=10**(0))
ax[0].grid(True, which="both", ls="-", color='0.65')
ax[0].legend()

ax[1].stem(length, error)
ax[1].set_xlabel("Number of iteration")
ax[1].set_ylabel('Absolute value of error')
#ax[1].set_xlim(left=10**(0))
ax[1].grid(True, which="both", ls="-", color='0.65')
```

* Use the Euler series to determine the value of $\pi$
$$ \frac{\pi^2}{6} = (\frac{1}{1^2}+\cfrac{1}{2^2}+\cfrac{1}{3^2}+\cfrac{1}{4^2}+\frac{1}{5^2}\cdots)$$

```python
n=5000
iteration=np.arange(1, n+1)
pi_array=np.zeros(n)
error=np.zeros(n)
s=0
for k in iteration:
    series=1/k**2
    s=s+series
    appro_pi=np.sqrt(6*s)
    pi_array[k-1]=appro_pi
    er=np.abs(np.pi-appro_pi)
    error[k-1]=er
length=np.arange(1, len(pi_array)+1)

fig, ax=plt.subplots(nrows=2, ncols=1, figsize=(15,10))

ax[0].semilogx(length, pi_array, 
               color="r", 
               linewidth=2, 
               label=r"Approximated $\pi$")
ax[0].plot([1, length[-1]], [np.pi, np.pi], 
           linestyle="--", 
           linewidth=2, 
           label=r"$\pi$")
ax[0].set_xlabel("Number of iteration in log scale")
ax[0].set_ylabel(r"Approximated value of $\pi$")
ax[0].set_xlim(left=10**(0))
ax[0].grid(True, which="both", ls="-", color='0.65')
ax[0].legend()

ax[1].semilogx(length, error, 
               color="DarkBlue", 
               linewidth=2)
ax[1].set_xlabel("Number of iteration in log scale")
ax[1].set_ylabel('Absolute value of error')
ax[1].set_xlim(left=10**(0))
ax[1].grid(True, which="both", ls="-", color='0.65')
```

* Use the Rieman zeta function $\zeta(2)$ to determine the value of $\pi$
$$ \frac{\pi^2}{6} = (\frac{2^2}{2^2-1}\cdot\cfrac{3^2}{3^2-1}\cdot\cfrac{5^2}{5^2-1}\cdot\cfrac{7^2}{7^2-1}\cdot\frac{11^2}{11^2-1}\cdots)$$

```python
def primes_upto(limit):
    prime = [True] * limit
    for n in range(2, limit):
        if prime[n]:
            yield n # n is a prime
            for c in range(n*n, limit, n):
                prime[c] = False # mark composites

primes=np.array(list(primes_upto(10000)))

pi_array=np.zeros(len(primes))
error=np.zeros(len(primes))
count=0
multiply=1
for k in primes:
    series=k**2/(k**2-1)
    m=series
    multiply=multiply*m
    appro_pi=np.sqrt(6*multiply)
    er=abs(np.pi-appro_pi)
    count=count+1
    pi_array[count-1]=appro_pi
    error[count-1]=er

length=np.arange(1, len(primes)+1)
fig, ax=plt.subplots(nrows=2, ncols=1, figsize=(15,10))

ax[0].semilogx(length, pi_array, 
               color="r", 
               linewidth=2, 
               label=r"Approximated $\pi$")
ax[0].plot([1, length[-1]], [np.pi, np.pi], 
           linestyle="--", 
           linewidth=1.5, 
           label=r"$\pi$")
ax[0].set_xlabel("Number of iteration in log scale")
ax[0].set_ylabel(r"Approximated value of $\pi$")
ax[0].set_xlim(left=10**(0))
ax[0].grid(True, which="both", ls="-", color='0.65')
ax[0].legend()
ax[1].semilogx(length, error, 
               color="DarkBlue", 
               linewidth=2)
ax[1].set_xlabel("Number of iteration in log scale")
ax[1].set_ylabel('Absolute value of error')
ax[1].set_xlim(left=10**(0))
ax[1].grid(True, which="both", ls="-", color='0.65')
```
