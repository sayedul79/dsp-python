```python
import numpy as np
import scipy.signal as sig
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display
```

```python
sample=10
f=10  #Maximum frequency of signal
f_s=40
T_s=1/f_s  
n=np.arange(0,sample)
nT=n*T_s
#sig=np.sin(2*np.pi*5*nT)
sig=np.sin(2*np.pi*n/sample)
#Reconstruction of digital signal
t=np.arange(0, sample, 0.1)
length=np.arange(len(n))
ct=0
fig, ax=plt.subplots(nrows=2, ncols=1, figsize=(12, 8))
for k in length:
    ct=sig[k]*np.sinc((t-k))
    ax[0].stem(n, sig)
    ax[1].plot(t, ct, color="green")
```

```python
def SamplingReconstruction(f_s=20):
    sample=30
    f=10  #Maximum frequency of signal
    #f_s=30
    T_s=1/f_s  
    n=np.arange(0,sample, 0.1)
    nT=n*T_s
    sig=np.sin(2*np.pi*f*nT)
    #Reconstruction of digital signal
    t=np.arange(0, 30, 0.001)
    length=np.arange(len(n))
    ct=0
    for k in length:
        ct=ct+sig[k]*np.sinc((t-k))
    fig, ax=plt.subplots(nrows=3, ncols=1, figsize=(12, 8))
    ax[0].stem(n, sig)
    ax[1].plot(t, ct, color="green")
    ax[2].plot(t, np.sin(2*np.pi*f*t))
```

```python
f_s=widgets.IntSlider(min=1, 
                      max=100, 
                      step=2, 
                      description=r"$f_s$", 
                      value=30)
plot=widgets.interactive_output(SamplingReconstruction, 
                                {"f_s": f_s})
layout=widgets.VBox([f_s, plot])
display(layout)
```
