# Introduction 

This is some text with math.
\begin{align}
y[n] = \sum_{k=0}^\infty x[n-k]
\end{align}

Suppose that $x[n] = u[n]$, then
\begin{align}
    y[n] &= \sum_{k=0}^\infty u[n-k]
\end{align}
In the sum above make the variable substitution $m = n-k$. The sum limits for finite $n$ become $k=0\Rightarrow m = n-0=n$ and $k=\infty\Rightarrow m = n-\infty = -\infty$. Making substitutoins into the original sum, we have
\begin{align}
    y[n] = \sum_{m=n}^{-\infty} u[m] = \sum_{m=-\infty}^n u[m] =       
  \begin{cases}
    0, & n < 0 \\
    n+1, & n\geq 0
  \end{cases}
\end{align}
We see that as $n\rightarrow \infty$ $y[n]\rightarrow\infty$ making the system unstable.

## Difference Equations

````python
def pulse_resp(n,a,N):
    """
    A function to find the pulse response of a 1st-order system
    y = pulse_resp(n,a,N)
    n = an array of sample values
    a = is a parameter
    N = the pulse duration in samples
    """
    y = zeros(len(n))
    for k,nk in enumerate(n):
        if nk < 0:
            y[k] = 0
#         elif nk >= 0 and nk <= N-1:
        elif nk <= N-1:
            y[k] = (1 - a**(nk+1))/(1 - a)
        else:
            y[k] = a**(nk-N+1)*(1 - a**N)/(1 - a)
    return y
````

### Example 2.11 Part a
$$
    y[n] -3 y[n-1] - 4 y[n-2] = x[n] + 2x[n-1]
$$
A problem to watch out for with `numpy` ndarray operations is managing data types. In the calculation of the closed-form impulse response, we need to compute $6/5\cdot 4^n$. The narray `n` is an integer type and yes Python does suppose integer operation with infinite precision. Since we are raising a scalar to a power that is an array unexpected things can happen when the numbers become large. In this case it is better to make 4 a float, i.e., 4.0, or make `n` and float64 array, i.e., `float64(n)`. I chose the latter to solve the calulation problem we saw in class on 9/18/2019.

````python
n = arange(50)
x_d = ss.dimpulse(n)
y_d = signal.lfilter([1,2],[1,-3,-4],x_d)
y_d_cf = -1/5*(-1)**n + 6/5*(4)**float64(n)
````

````python
def plotly_stem(x,y,xlabel,ylabel,title,stem_symbol_color = 'blue',
               stem_color = 'blue', stem_symbol_size=6,stem_width=0.25,
               xlim=None,ylim=None):
    """
    Produce a Plotly stem plot of the sequence in y with integer valued x values
    using plotly primitives Scatter and Bar. The interface is similar to 
    pyplot.stem() of matplotlib
    
    x : ndarray of x-axis data points
    y : ndarray of y-axis data points
    xlabel : string label for the x-axis (should support LaTeX)
    ylabel : string label for the y-axis (should support LaTeX) 
    title :  string for the plot title (should support LaTeX)
    stem_symbol_color : string color, e.g. 'blue', 'orange', etc. or 'rgb(R,G,B)'
    line_color : string color, e.g. 'blue', 'orange', etc. or 'rgb(R,G,B)'
    stem_symbol_size : optional size of symbol at the stem top
    stem_width : width of the stem as a percentage of the integer axis 
    space normalize by the x-axis length 
    xlim : optional axis limit as a two element list = [xmin, xmax]
    ylim : optional axis limit as a two element list = [ymin, ymax]
    
    Returns
    =======
    fig : Plotly figure object
    
    Notes
    =====
    It should be possible to include LaTeX math expressions in labels:
    xlabel, ylabel, and title, but I am unable to get this to work on
    of macOS and Windows. The object fig contains the initial plot state
    and hence is used also for the saved SVG figure when notebook convert
    is used to export to say *.md or *.pdf via LaTeX
    
    Mark Wickert September 2019
    """
    fig = go.Figure(data=go.Scatter(x=x, y=y,
                    mode='markers', name='',
                    marker=dict(size=stem_symbol_size,
                                color=stem_symbol_color)))
    Dx = x[-1] - x[0]
    if xlim != None:
        Dx = xlim[1] - xlim[0]
    stem_width = stem_width*Dx/100
    fig.add_trace(go.Bar(x=x, y=y, width=stem_width,
                  name='', marker_color=stem_color,
                  showlegend=False))
    fig.update_layout(title=title,
                      xaxis_title=xlabel, 
                      yaxis_title=ylabel)
    if xlim != None:
        fig.update_layout(xaxis=dict(range=xlim))
    if ylim != None:
        fig.update_layout(yaxis=dict(range=ylim))
    fig.update_layout(showlegend=False)
    fig.show() 
    return fig
````

## Aliased Sinc Function
The sinc function is typically defined as
$$
    \text{sinc}(x) \equiv \frac{\sin(\pi x)}{\pi x}
$$
Notice also that the aliased sinc can be defined in terms of the sinc function:
$$
    \frac{\sin(\omega M/2)}{\sin(\omega/2)} = \frac{\sin(\omega M/2)}{\omega M/2}\cdot\frac{\omega M/2}{\sin(\omega/2)} = M\frac{\text{sinc}(\omega M/(2\pi)}{\text{sinc}(\omega/(2\pi)}
$$
This expansion also shows how the scale factor $M$ appears.
Writing in terms of sinc functions manages the singularity when the argument is zero.

````python
M = 4
w = arange(0,2*pi,.01)
W_s = sinc(M*w/2/pi)
plot(w/2/pi,M*abs(W_s))
# W_as = sin(w*M/2)/sin(w/2)
W_as = M*sinc(w*M/2/pi)/sinc(w/2/pi)
plot(w/2/pi,abs(W_as))
title(r'Consider M = {}'.format(M))
ylabel(r'Magnitude')
xlabel(r'Normalized Frequency ($\omega/(2\pi)$)')
grid();
````

## Frequency Response and Spectrum

Next we move on to frequency response and the spectrum by considering a simple rectangular window signal of the form
\begin{align}
    w[n] = u[n] - u[n-M]
\end{align}
In applications of spectral estimation we might also view the rectangle as a window function, that is perhaps we form a *windowed* signal $x_w[n]$ via
\begin{align}
    x_w[n] = x[n] \times w[n]
\end{align}
where now $w[n]$ is the rectangular pulse signal and $x[n]$ is a signal of interest being acquired from an analog signal $x(t)$ via an ADC.

For comparison purposes consider also a shaped window formed using `scipy.signal.windows`. To see the available window functions we can list them using

```python
dir(signal.windows)
```
Editing the list and placing it in this markdown cell, we have:
`barthann`, `bartlett`, `blackman`, `blackmanharris`, `bohman`, `boxcar`, `chebwin`, `cosine`, `flattop`, `gaussian`, `general_gaussian`, `hamming`,  `hann`, `hanning`, `kaiser`, `nuttall`, `parzen`, `slepian`, `triang`

````python
M = 50
w_rect = ones(M) # also signal.boxcar()
w_hann = signal.hann(M)
#List the values by first forming a tuple
(w_rect,w_hann)
````

````python
figure(figsize=(6,3))
n = arange(0,len(w_rect))
subplot(211)
stem(n,w_rect,use_line_collection=True)
xlabel(r'Index $n$')
ylabel(r'Amplitude')
title(r'Rectangle Window for $M = $%d' % M)
ylim([0,1.1])
grid();
subplot(212)
stem(n,w_hann,use_line_collection=True)
xlabel(r'Index $n$')
ylabel(r'Amplitude')
title(r'Hann (Hanning) Window for $M = $%d' % M)
ylim([0,1.1])
grid();
tight_layout()
````

### Consider Window the Frequency Response
Next plot the spectrum of each of the two windows:

````python
f = arange(-.5,.5,.001)
w,W_rect = signal.freqz(w_rect,[1],2*pi*f)
w,W_hann = signal.freqz(w_hann,[1],2*pi*f)
````

````python
figure(figsize=(6,3))
plot(f,20*log10(abs(W_rect)/sum(w_rect))) #normalize
plot(f,20*log10(abs(W_hann)/sum(w_hann))) #normalize
xlabel(r'Frequency $\omega/(2\pi)$ (cycles/samp)')
ylabel(r'$|W(je^{\omega})|$')
title(r'Window Function Spectra: $W_{rect}$ and $W_{hann}$, $M$ = %d' % M)
ylim([-60,0])
legend((r'Rect',r'Hann'),loc='best')
grid();
````

Note how rapidly the spectrum of the *hann* (or hanning) window drops off compared with the simple rectangle.

## Spectral Estimation and Windowing

A simple spectrum estimation example is to now define signal $x[n]$ to be
\begin{align}
    x[n] = \cos(\omega_0 n),\ -\infty < n < \infty
\end{align}
multiply by the window functions and then consider the DTFT of the results.

````python
n = arange(0,1000)
x = cos(2*pi*1/10*n)
f = arange(-1.5,1.5,.001)
w,X = signal.freqz(x,[1],2*pi*f)
plot(f,abs(X)/1000)
grid();
````

````python
#Form x
w0 = 2*pi/10
M = 100
n = arange(M)
x = cos(w0*n)+ 1/50*cos(1.5*w0*n)
w_rect = ones(M) # also signal.boxcar()
w_hann = signal.hann(M)
#Form x_w_rect and x_w_hann
x_w_rect = x*w_rect
x_w_hann = x*w_hann
````

Compute the DTFT using `freqz` with the signal values loaded into the `b` array and the `a` array just 1.

````python
f = arange(-.5,.5,.001)
w,X_w_rect = signal.freqz(x_w_rect,[1],2*pi*f)
w,X_w_hann = signal.freqz(x_w_hann,[1],2*pi*f)
````

Notice how much more dynamic range there is when using the windowing. The frequency resolution is still not very good, but the window size is still very small. Try increasing it to see what happends.

# Problem 2.44
Creating a function for manipulating $x[n]$. This allows you to answer parts (d) and (e) a little bit easier. As an example function consider
\begin{align}
    x[n] = (3+1j)\delta[n+1] + (2-3j)\delta[n] + (1+1j)\delta[n-1]
\end{align}

````python
def x_pulse(n):
    """
    Create a function consisting of three impulse functions that fire at 
    n = -1, 0, 1, and have complex coefficients.
    """
    x = (3+1j)*ss.dimpulse(n+1) + (2-3j)*ss.dimpulse(n) \
        + (1-1j)*ss.dimpulse(n-1)
    return x
````

#### Plot Some Special Cases
Here we plot some special cases of $x[n]$:
1. $\text{Re}\{x[n]\}$
2. $\text{Im}\{x[n-2]\}$
3. $\text{Re}\{x[-n]\}$
4. $\text{Im}\{x^\ast[n]\}$

````python
n = arange(-10,10)
subplot(221)
title(r'Real Part of $x[n]$')
stem(n,real(x_pulse(n)),use_line_collection=True)
grid();
subplot(222)
title(r'Imaginary Part of $x[n-2]$')
stem(n,imag(x_pulse(n-2)),use_line_collection=True)
grid();
subplot(223)
title(r'Real Part of $x[-n]$')
stem(n,real(x_pulse(-n)),use_line_collection=True)
grid();
subplot(224)
title(r'Imaginary Part of $x^\ast[n]$')
stem(n,imag(conj(x_pulse(n))),use_line_collection=True)
grid();
tight_layout()
````
