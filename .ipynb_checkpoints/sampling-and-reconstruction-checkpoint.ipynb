```python
import numpy as np
import scipy.signal as sig
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display
```

```python
f=100  #Maximum frequency of signal
f_s=2000
T_s=1/f_s
N=int(f_s/f)
n=np.arange(0,N)
nT=n*T_s
sig=np.sin(2*np.pi*f*n)
#sig=np.sin(2*np.pi*n/sample)
#Reconstruction of digital signal
t=np.arange(0, (N-1)*T_s, 0.01)
#length=np.arange(len(n))
length=n
ct=0
fig, ax=plt.subplots(nrows=2, ncols=1, figsize=(12, 8))
for k in length:
    ct=sig[k]*np.sinc((t-k))
    ax[0].stem(n, sig)
    ax[1].plot(t, ct, color="green")
```

```python
SamplingFreq=1500
SamplingTime=1/SamplingFreq
SignalFreq=100
SampleNum=int(SamplingFreq/SignalFreq)
t=np.linspace(0, (SampleNum-1)*SamplingTime, SampleNum)
n=np.arange(0, SampleNum)
signal=np.sin(2*np.pi*SignalFreq*t)
# plt.stem(n, signal)
ct=0
print(list(enumerate(t)))
fig, ax=plt.subplots(nrows=2, ncols=1, figsize=(12, 8))
for index, shift in enumerate(t):
    t=np.arange(0, SampleNum, 0.01)
    ct=ct+signal[index]*np.sinc((t-index))
    ax[0].stem(n, signal)
    ax[1].plot(t, ct, color="green")
```

```python
def SamplingReconstruction(f_s=20):
    sample=30
    f=10  #Maximum frequency of signal
    #f_s=30
    T_s=1/f_s  
    n=np.arange(0,sample, 0.1)
    nT=n*T_s
    sig=np.sin(2*np.pi*f*nT)
    #Reconstruction of digital signal
    t=np.arange(0, 30, 0.001)
    length=np.arange(len(n))
    ct=0
    for k in length:
        ct=ct+sig[k]*np.sinc((t-k))
    fig, ax=plt.subplots(nrows=3, ncols=1, figsize=(12, 8))
    ax[0].stem(n, sig)
    ax[1].plot(t, ct, color="green")
    ax[2].plot(t, np.sin(2*np.pi*f*t))
```

```python
f_s=widgets.IntSlider(min=1, 
                      max=100, 
                      step=2, 
                      description=r"$f_s$", 
                      value=30)
plot=widgets.interactive_output(SamplingReconstruction, 
                                {"f_s": f_s})
layout=widgets.VBox([f_s, plot])
display(layout)
```
